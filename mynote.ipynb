```python

image = cv2.imread('image.jpg',cv2.IMREAD_GRAYSCALE)

minima= np.min(image)
maxima= 200

img_est= np.uint8(((image - minima)/(maxima - minima))*255)
hist_est= cv2.calcHist([img_est],[0],None,[256],[0,256])
  
# plotting the histogram
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.xlabel('intensity value')
plt.ylabel('number of pixels')
plt.plot(image)
plt.title('imagen original')


plt.subplot(2,2,1)
plt.imshow(img_est)
plt.title('stretched image')


plt.subplot(2,2,2)
plt.xlabel('intensity value')
plt.ylabel('number of pixels')
plt.plot(hist_est)
plt.title('Histogram of the stretched image')
plt.xlim([0,256])
```

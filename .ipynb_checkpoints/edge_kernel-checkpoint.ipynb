```python
import matplotlib.pyplot as plt
import cv2
import numpy
```

```python
img1= cv2.imread('image.jpg',cv2.IMREAD_COLOR)
imgBN= cv2.imread('image.jpg',cv2.IMREAD_GRAYSCALE)
imgdes= cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
cv2.imwrite('image.jpg',imgdes)

edge_kernel= numpy.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
imagen= cv2.filter2D(imgBN,-1,edge_kernel)

gaussian_kernel3= cv2.getGaussianKernel(500,0)
img2=cv2.filter2D(img1,-1,gaussian_kernel3) 
```

```python
plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.title('imagen original')
plt.imshow(img1)
plt.axis('off')

plt.subplot(1,3,2)
plt.title('deteccion de bordes')
plt.imshow(imagen, cmap='gray')
plt.axis('off')


plt.subplot(1,3,3)
plt.title('Desenfoque')
plt.imshow(img2)
plt.axis('off')

plt.tight_layout()
plt.show()
```

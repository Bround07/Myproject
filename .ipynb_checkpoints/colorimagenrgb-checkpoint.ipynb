```python
import cv2
import matplotlib.pyplot as plt
import cmapy

# Read image.
img = cv2.imread('image.jpg')

# Colorize.
img_colorizedr = cv2.applyColorMap(img, cmapy.cmap('PuBu'))
img_colorizedb = cv2.applyColorMap(img, cmapy.cmap('winter'))
img_colorizedg = cv2.applyColorMap(img, cmapy.cmap('BuGn_r'))

# Display
plt.imshow(img_colorizedr)
plt.show()
plt.imshow(img_colorizedb)
plt.show()
plt.imshow(img_colorizedg)
plt.show()
```

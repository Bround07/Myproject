```python
import numpy
import cv2
from matplotlib import pyplot as plt
import cmapy
```

```python
image = cv2.imread('image.jpg')
image2 = cv2.imread('meme.jpg')

imgm = cv2.imshow("image1",image)

b,g,r=cv2.split(image)
# cv2.imshow("momia", image)cv2.imshow("b", b)cv2.imshow("g", g)cv2.imshow("r", r)
```

```python
imgr = cv2.imshow("immgm",r)
cv2.imshow("hoa",b)
cv2.imshow("bob",g)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python


# Colorize.
img_colorizedr = cv2.applyColorMap(image, cmapy.cmap('PuBu'))
img_colorizedb = cv2.applyColorMap(image, cmapy.cmap('winter'))
img_colorizedg = cv2.applyColorMap(image, cmapy.cmap('BuGn_r'))

# Display
plt.imshow(img_colorizedr)
plt.show()
plt.imshow(img_colorizedb)
plt.show()
plt.imshow(img_colorizedg)
plt.show()
```

```python
histogram = cv2.calcHist(b, [0] , None , [256], [0, 256])
```

```python
histogram2 = cv2.calcHist(g, [0] , None, [256], [0, 256])
```

```python
histogram3 = cv2.calcHist(r, [0] , None, [256], [0, 256])
```

```python
hist= cv2.calcHist([r,g,b],[0],None,[256],[0,256])
colors= ('b','g','r')
```

```python
plt.figure()
plt.title('Histograma en escala rgb')
plt.xlabel('Bins')
plt.ylabel('# de pixeles')
plt.xlim((0, 256))
```

## Histograma blue

```python
# Aadimos el histograma al grfico.
plt.plot(histogram)

# Mostramos el grfico.
plt.show()
```

## Histograma green

```python
plt.plot(histogram2, color="g")

# Mostramos el grfico.
plt.show()
```

## Histograma red

```python
plt.plot(histogram3, color="r")

# Mostramos el grfico.
plt.show()
```

## histograma rgb

```python
plt.figure()
plt.title('Histograma en escala rgb')
plt.xlabel('Bins')
plt.ylabel('# de pixeles')
plt.xlim((0, 256))
for (i,col) in zip([b,g,r],colors):
    hist = cv2.calcHist([i],[0],None,[256],[0,256])    
    plt.plot(hist, color =col)
    plt.xlim([0,256])
```

```python
plt.close()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
